# Tic Tac Toe   

 

```python
import numpy as np
```

```python
# TicTacToe
board = np.full((3,3),' ')
board

def user_input(row, col, symbol):
    if symbol in ['X', 'O'] and 0 <= row < 3 and 0 <= col < 3:
        if board[row, col] == ' ':
            board[row, col] = symbol
        else:
            print("Invalid move")



```

```python
def check_winner():
    for i in range(3):
        if np.all(board[i, :] == 'X'): # all.() means all the elements in the row or column 
            return 'X wins!'
        elif np.all(board[:, i] == 'X'):
            return 'X wins!'
        elif np.all(board[i, :] == 'O') :
            return 'O wins!'
        elif np.all(board[:, i] == 'O'):
            return 'O wins!'


    flipp_board = np.fliplr(board)     
    
    if np.all(np.diag(board) == 'X'):# np.diag(board) check maian diagonl
        return 'X wins!'

    elif np.all(np.diag(board) == 'O'):
        return 'O wins!'

    elif np.all(np.diag(flipp_board) == 'X'): 
        return 'X wins!'
    elif np.all(np.diag(flipp_board) == 'O'):
        return 'O wins!'
    
    return None
```

```python
def print_board(board):
    print()
    for i in range(3):
        print(" " + " | ".join(board[i]))
        if i < 2:
            print("---+---+---")
    print()
```

```python
def play_game():

    current_player = "X"

    while True:

        print_board(board)
        print(f"Player {current_player}'s turn")

        # Input validation
        while True:
            try:
                row = int(input("Enter row (0-2): "))
                col = int(input("Enter column (0-2): "))

                if row not in range(3) or col not in range(3):
                    print("Invalid position. Choose numbers between 0 and 2.")
                    continue

                if board[row, col] != ' ':
                    print("Cell already taken.")
                    continue

                break

            except ValueError:
                print("Invalid input. Enter numbers only.")

        user_input(row, col, current_player)

        winner = check_winner()

        if winner:
            print(winner)
            print_board(board)
            break

        if np.all(board != ' '):
            print("Game Over - Draw")
            break

        current_player = "O" if current_player == "X" else "X"

        print("----------------------")
```

```python
# start the game
play_game()
```
